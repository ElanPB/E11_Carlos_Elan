```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../raspberry_files/CSV_FILES/20230215-160202_aq_pth.csv')

plt.plot(df['Unix Time'], df['PM1.0'])
plt.plot(df['Unix Time'], df['PM2.5'])
plt.plot(df['Unix Time'], df['PM10'])
plt.show()

plt.subplot(2, 3, 1)
plt.plot(df['Unix Time'], df['Temperature'], label = 'Temperature')
plt.legend(loc='best')
plt.subplot(2, 3, 2)
plt.plot(df['Unix Time'], df['Gas'], label = 'Gas')
plt.legend(loc='best')
plt.subplot(2, 3, 3)
plt.plot(df['Unix Time'], df['Humidity'], label = 'Humidity')
plt.legend(loc='best')
plt.subplot(2, 3, 4)
plt.plot(df['Unix Time'], df['Pressure'], label = 'Pressure')
plt.legend(loc='best')
plt.subplot(2, 3, 5)
plt.plot(df['Unix Time'], df['Altitude'], label = 'Altitude')
plt.legend(loc='best')
plt.show()
```
